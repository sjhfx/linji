```python
# 优化 Markdown 文档格式

import os
import re

def optimize_markdown(content):

    # 优化格式
    content = content.replace("  **1\.\n", "\n\n**1.")
    content = content.replace("  **2\.\n", "\n\n**2.")
    content = content.replace("  **3\.\n", "\n\n**3.")

    content = content.replace("  **1\. ", "\n\n**1.")
    content = content.replace("  **2\. ", "\n\n**2.")
    content = content.replace("  **3\. ", "\n\n**3.")
    
    content = content.replace("**1\.", "**1.")
    content = content.replace("**2\.", "**2.")
    content = content.replace("**3\.", "**3.")
    
    content = content.replace(" **① ", "\n\n**①")
    content = content.replace(" **② ", "\n\n**②")
    content = content.replace(" **③ ", "\n\n**③")
    content = content.replace(" **④ ", "\n\n**④")
    content = content.replace(" **⑤ ", "\n\n**⑤")
    content = content.replace(" **⑥ ", "\n\n**⑥")
    content = content.replace(" **⑦ ", "\n\n**⑦")
    content = content.replace(" **⑧ ", "\n\n**⑧")
    content = content.replace("**\n：", "**：")
    #content = content.replace("  ", "\n\n")
    content = content.replace("\n模型？**", " 模型？**")
    content = content.replace("\n 模型？**", " 模型？**")
    content = content.replace("模型？** ", " 模型？**\n")
    
    content = content.replace("  [", " [")
    content = content.replace("[ ", "[")
    content = content.replace("\n]", "]")

    content = content.replace("**最后的话**", "\n\n**最后的话**\n\n")
    sentences = re.split(r'(?<![。]\*\*)(?<=[？]\*\*)', content)
    content = '\n\n'.join(sentences)
    # 去掉多余的空行
    content = re.sub(r'(\d+|[a-zA-Z]+|「|\))\n', r'\1 ', content)
    # 确保段落之间有空行
    content = re.sub(r'\n{3,}', '\n\n', content)
    content = re.sub(r'(?<!\n)\n(?!\n)', '\n\n', content)

    # 将每个完整的句子作为单独的段落
    #sentences = re.split(r'(?<=[.!?])\s+', content)
    #content = '\n\n'.join(sentences)
    
    return content

def process_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    
    optimized_content = optimize_markdown(content)
    
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(optimized_content)

def batch_process_markdown(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.md'):
                file_path = os.path.join(root, file)
                process_file(file_path)
                print(f"已优化: {file_path}")


target_directory = "d:/linjipro/linji/tech/model"
batch_process_markdown(target_directory)
```
